```python
#Outlier Detection for all columns
for column_name in numeric_columns:
      print("Outliers calculation for :", column_name)
      print(sale_data[column_name].describe())
      quartile1 = sale_data[column_name].quantile(0.25)
      quartile3 = sale_data[column_name].quantile(0.75)
      print('Quartile 1 : ', quartile1,'and Quartile 3 : ', quartile3)
      IQR = quartile3 - quartile1
      print('Inter quartile Range: ',IQR)
      #Upper Limit
      UL = quartile3 + (1.5 * IQR)
      print('Upper Limit : ',UL)
      #Lower Limit
      LL = quartile1 - (1.5 * IQR)
      print('Lower Limit : ', LL)
      print('The Outliers for column',column_name, 'are:')
      Outliers=sale_data[(sale_data[column_name] > UL) | (sale_data[column_name] < LL)]
      print(Outliers)
      print('--------------------------------------------------------') 
```

### Standardization of Numerical Variables

```python
sale_numerical_data = sale_data[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']]
sale_numerical_data
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
np.set_printoptions(precision=2,suppress=True)
scaler=StandardScaler()
rescaledX=scaler.fit_transform(sale_numerical_data)
data1 = pd.DataFrame(rescaledX, columns=list(numeric_columns))
threshold = 2
for column_name in data1:
    print("Outliers calculation for :", column_name,'Threshold ' ,threshold)
    c_array = np.array(data1[column_name])
    c_data = np.array(sale_data[column_name])
    z_scores = c_array.flatten()
    outliers = c_data[np.abs(z_scores) > threshold]
    print("Outliers:", outliers)
    print("---------------------------------------------------------------------")
```
