```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('D:\\Data-science\\anime.csv')
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data.info()
```

```python
data['episodes']  = data['episodes'].apply(pd.to_numeric, errors='coerce')
median_ep = data['episodes'].median()
data['episodes'] = data['episodes'].fillna(median_ep)
data.info()
```

```python
#number of unique users in the dataset
len(data.name.unique())
```

```python
df_genres = data['genre'].str.get_dummies(sep=',').add_prefix('genre_')
df_genres
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['episodes', 'rating', 'members']] = scaler.fit_transform(data[['episodes', 'rating', 'members']])

# Merge the one-hot encoded genres with the original DataFrame
df = pd.concat([data, df_genres], axis=1)

# Display the processed DataFrame
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
features = df.drop(columns=["anime_id", "name", "type","genre"])
cosine_sim = cosine_similarity(features)
```

```python
def get_recommendations(title, cosine_sim, df, threshold):
    idx = df[df['name'] == title].index[0]
    print(idx)
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = [score for score in sim_scores if score[1] >= threshold]

    sim_scores = sim_scores[1:6]
    print(sim_scores)
    anime_indices = [i[0] for i in sim_scores]
    return df[['name','genre']].iloc[anime_indices]

recommendations = get_recommendations('Ginga Eiyuu Densetsu', cosine_sim, df, 0.1)
recommendations
```

#### Can you explain the difference between user-based and item-based collaborative filtering?

<div><b>User-based and Item-based Collaborative Filtering are two popular techniques used in collaborative filtering for recommendation systems. Both approaches rely on past interactions (like ratings or purchases) to make predictions about what a user might like in the future</b></div>
<div><span>User-based Collaborative Filtering:</span> 
<span>User-based collaborative filtering focuses on recommending items to a user based on the preferences of other similar users. The idea is that if User A and User B have similar preferences or behavior in the past, then they are likely to share similar tastes in the future</span>
<span>Common similarity measures include cosine similarity, Pearson correlation, or Euclidean distance.</span>
<ui>Advantages: 
<li>Simple to understand and implement.</li>
<li>Can provide good recommendations when the user’s past behavior aligns with other users' behavior.
</li>
<ui>
<ui>Disadvantages
<li>Scalability issues</li>
<li>Cold start problem</li>
<li>Sparsity problem</li>
<ui>
<div>
<br/>
<div><b>Item-based collaborative filtering focuses on recommending items based on the similarity between items themselves, rather than the similarity between users. </b>
<span>Calculate the similarity between items, often using cosine similarity or Pearson correlation between item rating vectors</span><br/>
<ui>Advantages: 
<li>Scalability</li>
<li>Less affected by the cold start problem</li>
</ui>
<ui>Disadvantages:
<li>Cold start problem for new items</li>
<li>Limited personalization</li>
<ui>
</div>


### What is collaborative filtering, and how does it work?

<div><b>Collaborative filtering is a recommendation technique that predicts a user's preferences for items (like movies, products, or songs) based on the preferences of other users. </b></div>
<ui>1. User-Based Collaborative Filtering:
<li>Recommend items to a user based on what similar users liked</li>
<li>Ex: If Alice and Bob both liked “Inception” and “Interstellar,” and Bob also liked “The Matrix,” then Alice might be recommended “The Matrix.”<li>
</ui>
<ui>2. Item-Based Collaborative Filtering
<li>Recommend items similar to the ones the user already liked</li>
<li>Ex: If many users who liked “Inception” also liked “The Matrix,” then a user who liked “Inception” might be recommended The Matrix.</li>
</ui>
