```python
data.replace(" ?", pd.NA, inplace=True)
data.isna().sum()
```

```python
for col in data.select_dtypes(include='object').columns:
    data[col].fillna(data[col].mode()[0], inplace=True)
data
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
num_cols = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
standard_scaled = pd.DataFrame(StandardScaler().fit_transform(data[num_cols]), columns=num_cols)
minmax_scaled = pd.DataFrame(MinMaxScaler().fit_transform(data[num_cols]), columns=num_cols)
```

```python
from sklearn.preprocessing import LabelEncoder
# Separate categorical columns
cat_cols = data.select_dtypes(include='object').columns.drop('income')

# One-Hot for < 5 categories
one_hot_cols = [col for col in cat_cols if data[col].nunique() < 5]
print(one_hot_cols)

data = pd.get_dummies(data, columns=one_hot_cols, drop_first=True)
print(data)
# Label Encoding for high-cardinality
label_cols = [col for col in cat_cols if data[col].nunique() >= 5]
print(label_cols)
le = LabelEncoder()
for col in label_cols:
    data[col] = le.fit_transform(data[col])
```

```python
data['sex'] = data['sex_ Male'].replace({True:1, False:0})
```

```python
data=data.drop(['sex_ Male'], axis=1)
```

```python
data
```

```python
# Feature Engineering
# Age group
data['age_group'] = pd.cut(data['age'], bins=[0, 25, 45, 65, 100], labels=['Young', 'Adult', 'Middle', 'Senior'])

# Working hour category
data['hours_category'] = pd.cut(data['hours_per_week'], bins=[0, 20, 40, 60, 100], labels=['Very Low', 'Low', 'Medium', 'High'])
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Before transform
sns.histplot(data['capital_gain'], kde=True)

# Log transform
data['capital_gain_log'] = np.log1p(data['capital_gain'])

# After transform
sns.histplot(data['capital_gain_log'], kde=True)
```

```python
from sklearn.ensemble import IsolationForest
iso = IsolationForest(contamination=0.01, random_state=42)
outlier_flags = iso.fit_predict(data[num_cols])
data = data[outlier_flags == 1]  # Remove outliers
data
```

```python
import ppscore as pps
# PPS matrix
pps_matrix = pps.matrix(data)
pps_heatmap = pps_matrix.pivot(index='x', columns='y', values='ppscore')

# Visualize
plt.figure(figsize=(12, 10))
sns.heatmap(pps_heatmap, cmap='coolwarm', annot=False)
```

```python
# Pearson correlation (only numeric)
numeric_df = data.select_dtypes(include='number')
corr_matrix = numeric_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
```

<b>PPS captures non-linear relationships, even between categorical and numerical variables.</b>
<br/><b>Correlation is only suitable for numeric variables and linear dependencies.</b>
